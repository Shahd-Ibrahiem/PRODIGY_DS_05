```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#Import Traffic Dataset

```python
df = pd.read_csv("Traffic.csv")
```

#Preprocessing Data

```python
df.info()
```

```python
df.head()
```

```python
df.isnull().sum()
```

#Visualizing Accident Distribution by Road Surfaces

```python
plt.figure(figsize=(10 , 10))
sns.countplot(x="Road_surface_type", data = df)
plt.xticks(rotation=45)
plt.xlabel("Road Surface Type")
plt.ylabel("Number of Accidents")
plt.title("Accidents by Road Surface Type")
plt.show()
```

#Visualizing Accident Distribution by Years of Driving Experience

```python
plt.figure(figsize=(10 , 10))
sns.countplot(x="Driving_experience", data = df)
plt.xticks(rotation=45)
plt.xlabel("Driving_experience")
plt.ylabel("Number of Accidents")
plt.title("Accidents by Years of Driving experience")
plt.show()
```

#Visualizing Accident Distribution by Weather Conditions

```python
plt.figure(figsize=(10 , 10))
sns.countplot(x="Weather_conditions", data = df)
plt.xticks(rotation=45)
plt.xlabel("Weather Conditions")
plt.ylabel("Number of Accidents")
plt.title("Accidents by Weather Conditions")
plt.show()
```

#Visualizing Accident Distribution by Accident Severity

```python
plt.figure(figsize=(10 , 10))
sns.countplot(x="Accident_severity", data = df)
plt.xlabel("Accident Severity")
plt.ylabel("Number of Accidents")
plt.title("Accidents by Accident Severity")
plt.show()
     
```

#Visualizing Accident Distribution by Cause Of Accident

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Cause_of_accident', data = df)
plt.xticks(rotation=90)
plt.xlabel("Cause of Accident")
plt.ylabel("Number of Accidents")
plt.title("Accidents by Contributing Factors")
plt.show()
```

```python
#Visualizing Accident Severity by Cause Of Accident
```

```python
plt.figure(figsize=(10 , 10))
sns.countplot(x="Accident_severity", hue = "Cause_of_accident", data = df)
plt.xlabel("Accident Severity")
plt.ylabel("Number of Accidents")
plt.title("Accidents by Accident Severity")
plt.show()
```
